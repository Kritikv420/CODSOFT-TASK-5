```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.metrics import accuracy_score as acc_score
```

```python
cc_data = pd.read_csv('/content/creditcard.csv')
```

```python
legit_transactions = cc_data[cc_data['Class'] == 0]
fraud_transactions = cc_data[cc_data['Class'] == 1]
```

```python
print(legit_transactions.shape)
print(fraud_transactions.shape)
```

```python
legit_transactions['Amount'].describe()
```

```python
fraud_transactions['Amount'].describe()
```

```python
cc_data.groupby('Class').mean()
```

```python
sampled_legit = legit_transactions.sample(n=492)
```

```python
combined_dataset = pd.concat([sampled_legit, fraud_transactions], axis=0)
```

```python
combined_dataset.head()
```

```python
combined_dataset.tail()

```

```python
combined_dataset['Class'].value_counts()
```

```python
combined_dataset.groupby('Class').mean()
```

```python
X = combined_dataset.drop(columns='Class', axis=1)
y = combined_dataset['Class']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train) # Changed Y_train to y_train
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test) # Changed Y_test to y_test
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
